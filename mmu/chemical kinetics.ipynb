Chemical Kinetics - conservation of mass
A -> B -> C

$\frac{dx}{dt}=-r_lx$

```python

import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
r1, r2, = 0.01, 0.02
x0, y0, z0 = 1000, 0, 0
tmax, n = 500, 500

def Chemical_Kinetics(X, t, r1, r2):
    x,y,z = X
    dx = -r1 * x
    dy = r1 * x - r2 * y
    dz = r2 * y
    return (dx, dy, dz)


t = np.linspace(0, tmax, n)
f = odeint(Chemical_Kinetics, (x0,y0,z0), t, args=(r1, r2))
x,y,z = f.T
plt.figure(1)
plt.xlabel("Time")
plt.ylabel("Concentrations")
plt.title("Chemical Kinetics")
plt.plot(t, x, label='|A|')
plt.plot(t, y, label='|B|')
plt.plot(t, z, label='|C|')
legend = plt.legend(loc='best')
plt.show()
```

```python
k, a0, b0, c0 = 0.00713, 4, 1, 0
def ode(c, t):
    dcdt = k * (a0 - c)**2 * (b0 - c / 2)
    return dcdt

t = np.linspace(0, 400, 401)
c = odeint(ode, c0, t)
plt.xlabel('Time')
plt.ylabel('c(t)')
plt.title('Production of Nitrogen Dioxide')
plt.plot(t, c)
plt.show()

print('c(100)= ', c[100])
```

```python
line, stage = [0, 1], 5
def cantor(line, level = 0):
    plt.plot(line, [level, level], color = "b", lw = 2, solid_capstyle = "butt")
    if level < stage:
        segment = np.linspace(line[0], line[1], 4)
        cantor(segment[:2], level + 1)
        cantor(segment[2:], level + 1)

cantor(line)
plt.gca().invert_yaxis()
plt.xticks([0, 1/3, 2/3, 1],['0', '1/3', '2/3', '1'])
plt.show()
```
